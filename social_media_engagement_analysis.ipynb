```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('social_media_data.csv')

# Hitung engagement rate
df['engagement_rate'] = (df['likes'] + df['comments'] + df['shares']) / df['likes']

# Lihat data
print(df.head())

```

```python

# ANALISIS 1: Engagement by Post Type
engagement_by_type = df.groupby('post_type')['engagement_rate'].mean()

print("\nAverage Engagement by Post Type:")
print(engagement_by_type)

engagement_by_type.plot(kind='bar')
plt.title('Average Engagement Rate by Post Type')
plt.xlabel('Post Type')
plt.ylabel('Engagement Rate')
plt.show()
```

```python
# ANALISIS 2: Best Posting Time
engagement_by_time = df.groupby('post_time')['engagement_rate'].mean()

print("\nAverage Engagement by Posting Time:")
print(engagement_by_time)

engagement_by_time.plot(kind='line', marker='o')
plt.title('Engagement Rate by Posting Time')
plt.xlabel('Hour')
plt.ylabel('Engagement Rate')
plt.show()
```

Based on the analysis of social media engagement data, several important insights were identified:

1. Content Type Performance

Video content shows the highest engagement rate compared to image and carousel posts. This indicates that audiences are more likely to interact with dynamic and visually engaging content.

Interpretation:
Users tend to prefer content that is more interactive and entertaining, making video a more effective format for increasing engagement.

2. Optimal Posting Time

Posts published during evening hours (around 19:00–21:00) achieve higher engagement rates than posts published during other times of the day.

Interpretation:
This suggests that audiences are more active on social media during the evening, likely after completing daily activities such as work or study.

3. Engagement Pattern Consistency

Content with consistently higher interaction (likes, comments, shares) contributes to a stronger overall engagement rate.

Interpretation:
Maintaining consistency in posting and content quality is crucial for sustaining audience interest and interaction over time.
